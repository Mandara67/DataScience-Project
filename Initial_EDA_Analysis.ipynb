```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/flights_sample_100k.csv')

# Quick look
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Drop rows where delay is missing (needed for analysis)
df = df.dropna(subset=['DEP_DELAY'])

# To keep delay under 5 hours tracked to avoid outliers
df = df[df['DEP_DELAY'] < 300]

# Convert date
df['FL_DATE'] = pd.to_datetime(df['FL_DATE'])

# Create useful features
df['MONTH'] = df['FL_DATE'].dt.month
df['DAY_OF_WEEK'] = df['FL_DATE'].dt.dayofweek
df['DEP_HOUR'] = df['CRS_DEP_TIME'] // 100
```

```python
df['DELAYED'] = (df['DEP_DELAY'] > 15).astype(int)
```

```python
plt.figure()
df['DEP_DELAY'].hist(bins=50)
plt.title("Distribution of Departure Delays")
plt.xlabel("Delay (minutes)")
plt.ylabel("Number of Flights")
plt.figure(figsize=(10,5))
plt.show()
```

```python
delay_rate = df['DELAYED'].mean()
print("Delay Rate:", delay_rate)
print("Delay Percentage: ", delay_rate*100, "%")
```

```python
plt.figure()
df.groupby('AIRLINE')['DEP_DELAY'].mean().sort_values().plot(kind='bar')
plt.title("Average Delay by Airline")
plt.xlabel("Airline")
plt.xticks(rotation=45, ha='right')
plt.ylabel("Avg Delay (minutes)")
plt.show()
```

```python
plt.figure()
df.groupby('DEP_HOUR')['DEP_DELAY'].mean().plot()
plt.title("Average Delay by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Avg Delay")
plt.show()
```

```python
plt.figure()
df['CANCELLED'].value_counts().plot(kind='bar')
plt.title("Cancelled vs Not Cancelled")
plt.show()

print("Cancellation Rate:", df['CANCELLED'].mean())
```

```python
route_delay = df.groupby(['ORIGIN', 'DEST'])['DEP_DELAY'].mean().reset_index()
route_delay.sort_values(by='DEP_DELAY', ascending=False).head(10)
```

```python
print("Average delay:", df['DEP_DELAY'].mean())
print("Max delay:", df['DEP_DELAY'].max())
print("Most common airline:", df['AIRLINE'].mode()[0])
```
